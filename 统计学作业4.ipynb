```python
import numpy as np
from scipy import stats

print("假设：H0>15%, H1<=15%")

n = 15
x_bar = 0.10
s = 0.06
mu_0 = 0.15

t_statistic = (x_bar - mu_0) / (s / np.sqrt(n))

df = n - 1

p_value = 1 - stats.t.cdf(t_statistic, df)

alpha = 0.05

print(f"样本大小 (n): {n}")
print(f"平均值 (x_bar): {x_bar:.2%}")
print(f"标准差 (s): {s:.2%}")
print(f"假设平均收益率 (mu_0): {mu_0:.2%}")
print(f"自由度 (df): {df}")
print(f"计算得到的t统计量: {t_statistic:.4f}")
print(f"P-值 (双尾): {p_value:.4f}")
print(f"显著性水平 (alpha): {alpha:.2f}")

if p_value < alpha:
    print("\n结论: 由于p值小于显著性水平，我们拒绝原假设")
    print("有充分证据表明平均收益15%以上，即股评家的预测正确。")
else:
    print("\n结论: 由于p值大于显著性水平，我们不能拒绝原假设。")
    print("有充分证据表明平均收益15%以上，即股评家的预测不正确。")
```

```python
import numpy as np
from scipy import stats

print("假设：H0=5000, H1!=5000")

n = 100
x_bar = 4950
s = 150
mu_0 = 5000
alpha = 0.05

t_statistic = (x_bar - mu_0) / (s / np.sqrt(n))

df = n - 1

p_value = stats.t.sf(abs(t_statistic), df) * 2

print(f"样本大小 (n): {n}")
print(f"样本平均收入 (x_bar): {x_bar} 元")
print(f"样本标准差 (s): {s} 元")
print(f"假设总体平均收入 (mu_0): {mu_0} 元")
print(f"自由度 (df): {df}")
print(f"计算得到的t统计量: {t_statistic:.4f}")
print(f"P-值 (双尾): {p_value:.4f}")
print(f"显著性水平 (alpha): {alpha:.2f}")

if p_value < alpha:
    print("\n结论: 由于p值小于显著性水平，我们拒绝原假设。")
    print("有充分证据表明农户家庭的平均收入不等于5000元，即统计员的说法不正确。")
else:
    print("\n结论: 由于p值大于显著性水平，我们不能拒绝原假设。")
    print("没有充分证据表明农户家庭的平均收入不等于5000元，因此统计员的说法可能是正确的 (即平均收入为5000元)。")
```

```python
import numpy as np
from scipy import stats

print("假设：H0 P<=20%, H1 P>20%")

p_0 = 0.20
n = 100
x = 16
alpha = 0.05

p_hat = x / n


SE = np.sqrt(p_0 * (1 - p_0) / n)

z_statistic = (p_hat - p_0) / SE

p_value = stats.norm.cdf(z_statistic)

print(f"假设总体订阅率 (p_0): {p_0:.2%}")
print(f"样本大小 (n): {n}")
print(f"样本中订阅的家庭数 (x): {x}")
print(f"样本订阅率 (p_hat): {p_hat:.2%}")
print(f"计算得到的z统计量: {z_statistic:.4f}")
print(f"P-值 (左尾): {p_value:.4f}")
print(f"显著性水平 (alpha): {alpha:.2f}")

if p_value < alpha:
    print("\n结论: 由于p值小于显著性水平，我们拒绝原假设。")
    print("有充分证据表明订阅率显著降低。")
else:
    print("\n结论: 由于p值大于显著性水平，我们不能拒绝原假设。")
    print("没有充分证据表明订阅率显著降低。")
    print("因此，无法得出订阅率显著降低的结论。")
```
