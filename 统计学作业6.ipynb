```python
import pandas as pd

# 字段重命名
df = pd.read_csv('hw6ex1.csv')
df.columns = ['Month', 'Total_Output_Value', 'Employees', 'Production_Workers']

# 计算该年各季度的月平均总产值和全年的月平均总产值
df['Quarter'] = (df['Month'] - 1) // 3 + 1

quarterly_avg_output = df.groupby('Quarter')['Total_Output_Value'].mean()
print("\n每季度月均总产值 (万元):")
print(quarterly_avg_output)

yearly_avg_output = df['Total_Output_Value'].mean()
print(f"\n全年月均总产值 (万元): {yearly_avg_output:.2f}")

# (2)全年平均职工人数
employees_base = 230
annual_avg_employees_list = df['Employees']
annual_avg_employees_list[0] = (employees_base + df['Employees'][0]) / 2
annual_avg_employees = annual_avg_employees_list.mean()
print(f"\n(2) 全年平均职工人数: {annual_avg_employees:.2f} 人")

# (3) 计算该年月平均全员劳动生产率、全年平均全员劳动生产率和全年职工构成指标
total_annual_output_value = df['Total_Output_Value'].sum()
annual_labor_productivity = (total_annual_output_value / 12) / annual_avg_employees
print(f"\n(3) 月平均全员劳动生产率: {annual_labor_productivity:.4f} 万元/人")


annual_labor_productivity = total_annual_output_value / annual_avg_employees
print(f"\n全年平均全员劳动生产率: {annual_labor_productivity:.4f} 万元/人")

# Annual average number of production workers
product_workers_base = 160
annual_avg_product_workers_list = df['Production_Workers']
annual_avg_product_workers_list[0] = (product_workers_base + df['Production_Workers'][0]) / 2
annual_avg_production_workers = annual_avg_product_workers_list.mean()

annual_employee_structure_index = annual_avg_production_workers / annual_avg_employees
print(f"全年职工构成指标可以定为：①全年生产工人比重②全年非生产工人比重")
print(f"①全年生产工人比重: {annual_employee_structure_index * 100:.2f}%")
print(f"②全年非生产工人比重: {(1- annual_employee_structure_index) * 100:.2f}%")


# (4) 对企业上半年、下半年的变化作出分析评价
print("\n--- 对比分析 ---")
df_h1 = df[df['Month'] <= 6]
df_h2 = df[df['Month'] > 6]

# 上半年
avg_output_h1 = df_h1['Total_Output_Value'].mean()
print(f"上半年月平均生产总值: {avg_output_h1:.2f} 万元")

avg_employees_h1 = df_h1['Employees'].mean()
print(f"上半年平均职工数: {avg_employees_h1:.2f} 人")

total_output_h1 = df_h1['Total_Output_Value'].sum()
labor_productivity_h1 = total_output_h1 / avg_employees_h1 if avg_employees_h1 else 0
print(f"上半年月平均全员劳动生产率: {labor_productivity_h1:.2f} 万元/人")

# 下半年

avg_output_h2 = df_h2['Total_Output_Value'].mean()
print(f"下半年月平均生产总值: {avg_output_h2:.2f} 万元")

avg_employees_h2 = df_h2['Employees'].mean()
print(f"下半年平均职工数: {avg_employees_h2:.2f} 人")

total_output_h2 = df_h2['Total_Output_Value'].sum()
labor_productivity_h2 = total_output_h2 / avg_employees_h2 if avg_employees_h2 else 0
print(f"下半年月平均全员劳动生产率: {labor_productivity_h2:.2f} 万元/人")



print(f"\n平均生产总值: 上半年平均 {avg_output_h1:.2f}万元, 下半年平均 {avg_output_h2:.2f}万元. "
      f"({'增长' if avg_output_h2 > avg_output_h1 else '降低' if avg_output_h2 < avg_output_h1 else '相同'}) {abs(avg_output_h2 - avg_output_h1):.2f}万元.")

print(f"平均职工数: 上半年平均 {avg_employees_h1:.2f}人, 下半年平均 {avg_employees_h2:.2f}人. "
      f"({'增长' if avg_employees_h2 > avg_employees_h1 else '降低' if avg_employees_h2 < avg_employees_h1 else '相同'}) {abs(avg_employees_h2 - avg_employees_h1):.2f}人.")

print(f"平均全员劳动生产率: 上半年平均 {labor_productivity_h1:.2f}万元/人, 下半年平均 {labor_productivity_h2:.2f}万元/人. "
      f"({'增长' if labor_productivity_h2 > labor_productivity_h1 else '降低' if labor_productivity_h2 < labor_productivity_h1 else '相同'}) {abs(labor_productivity_h2 - labor_productivity_h1):.2f}万元/人.")
```
