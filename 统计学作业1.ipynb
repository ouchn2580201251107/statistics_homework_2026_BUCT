```python
# Exercise 1
# ignore warnings
import warnings
warnings.filterwarnings("ignore")

import csv
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt

# homework 1 exercise 1
file_path = 'hw1ex1.csv'
df = pd.read_csv(file_path)

# Clean up column names by stripping whitespace and converting to lowercase
df.columns = df.columns.str.strip().str.lower()

df = df.apply(pd.to_numeric, errors='ignore')

# 0. 输出输入数据清单
print("输入数据为：")
print(df)

# 1. 该班级每月人均可支配收入的最大值与最小值
print("1. 该班级每月人均可支配收入的最大值与最小值")
total_income = df["income"]
max_income = max(total_income)
min_income = min(total_income)
print(f"该班级每月人均可支配收入的最大值为：{max_income}")
print(f"该班级每月人均可支配收入的最大值为：{min_income}")

# 2. 人均可支配收入的全距
print("2. 人均可支配收入的全距")
income_range = max_income - min_income
print(f"人均可支配收入的全距为：{income_range}")

# 3. 在人均可支配收入中以300元为间距，计算各区间收入的人数分布与相应的概率分布并绘制相应的柱状图；
bin_width = 300
bins_start = min_income
bins_end = math.ceil(income_range / bin_width) * bin_width + bins_start
bins = np.arange(bins_start, bins_end + 1, bin_width)

# 使用Pandas进行数据分箱处理
income_bins = pd.cut(df['income'], bins=bins, right=False, include_lowest=True)
income_count = income_bins.value_counts().sort_index()
income_ratio = (income_bins.value_counts(normalize=True).sort_index() * 100).apply(lambda x: f'{x:.1f}%')
result = pd.concat([income_count, income_ratio], axis = 1)
result.columns = ['Count', 'Ratio']

print("人均可支配收入的区间分布（频率）：")
print(result)

# 绘图
print("相应的柱状图")
plt.figure(figsize=(10, 6))
income_count.plot.bar(rot=45)
plt.title('Cost')
plt.xlabel('Income(yuan)')
plt.ylabel('Count')
plt.show()
```

```python
# Exercise 3
# ignore warnings
import warnings
warnings.filterwarnings("ignore")

import csv
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt

# homework 1 exercise 3
file_path = 'hw1ex3.csv'
df = pd.read_csv(file_path)

# Clean up column names by stripping whitespace and converting to lowercase
df.columns = df.columns.str.strip().str.lower()

print('(1).分析公司工人在性别方面人数分布情况')
count_gender = df.groupby('性别').size()
print(count_gender)

print('(2).分析公司工人在年龄方面人数分布情况')
bin_width = 10
bins_start = max(0, (math.ceil(min(df['年龄']) / bin_width) - 1)) * bin_width
bins_end = math.ceil(max(df['年龄'])/ bin_width)  * bin_width
bins = np.arange(bins_start, bins_end + 1, bin_width)
age_bins = pd.cut(df['年龄'], bins=bins, right=False, include_lowest=True)
age_count = age_bins.value_counts().sort_index()
print(age_count)

print('(3).分析公司工人在文化程度方面人数分布情况')
count_background = df.groupby('文化程度').size()
print(count_background)

print('(4).分析公司工人在技术级别方面人数分布情况')
count_level = df.groupby('技术级别').size()
print(count_level)
```

```python
# Exercise 4
# ignore warnings
import warnings
warnings.filterwarnings("ignore")

import csv
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt

# homework 1 exercise 4
file_path = 'hw1ex4.csv'
df = pd.read_csv(file_path)

# Clean up column names by stripping whitespace and converting to lowercase
df.columns = df.columns.str.strip().str.lower()

# 1. 该公司2024年度总销售额
sum_company = sum(df['公司销售额'])
print(f"该公司2024年度总销售额为：{sum_company}万元")

# 2. 该行业2024年度总销售额
sum_industry = sum(df['行业销售额'])
print(f"该行业2024年度总销售额为：{sum_industry}万元")

# 3. 图表展示（例如柱形图、散点图等）分析销售额时间变化趋势
print('结论：2024年企业销售额和行业销售额均整体呈现上升趋势。行业销售额增长相较于企业销售额更快。')
# 柱形图
print('柱形图')
plt.figure()
df.plot(kind='bar', figsize=(12, 7))
plt.title('Company vs. Industry Sal(es by Month in 2024')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend(['company_sales', 'industry_sales'])
plt.xticks(ticks = np.arange(12), labels = (np.arange(12) + 1).astype(int))
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()

# 散点图
print('散点图')
plt.figure(figsize=(12, 7))
plt.plot(df['月份'], df['公司销售额'], marker='o', linestyle='', color='b', label='company_sales')
plt.plot(df['月份'], df['行业销售额'], marker='x', linestyle='', color='r', label='industry_sales')
plt.xlabel('month')
plt.ylabel('sales')
plt.title('compare_point_chart')
plt.grid(True)
plt.xticks(df['月份'])
plt.legend()
plt.tight_layout()
plt.show()

# 折线图
print('折线图')
plt.figure(figsize=(12, 7))
plt.plot(df['月份'], df['公司销售额'], marker='o', linestyle='--', color='b', label='company_sales')
plt.plot(df['月份'], df['行业销售额'], marker='x', linestyle='-', color='r', label='industry_sales')
plt.xlabel('month')
plt.ylabel('sales')
plt.title('compare_line_chart')
plt.grid(True)
plt.xticks(df['月份'])
plt.legend()
plt.tight_layout()
plt.show()
```
