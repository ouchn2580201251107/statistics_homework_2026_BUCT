```python
import pandas as pd


df_raw = pd.read_csv('/content/hw7ex1.csv', encoding='utf-8')

col_mapping = {
  df_raw.columns[0]: 'Grain_Crop_Item',
  df_raw.columns[1]: 'Sown_Area_Base',
  df_raw.columns[2]: 'Sown_Area_Current',
  df_raw.columns[3]: 'Yield_Base',
  df_raw.columns[4]: 'Yield_Current',
  df_raw.columns[5]: 'Price_Base',
  df_raw.columns[6]: 'Price_Current'
}

# 数据处理
df = df_raw.rename(columns=col_mapping)
df_data = df.iloc[1:].copy()
numeric_cols = ['Sown_Area_Base', 'Sown_Area_Current', 'Yield_Base', 'Yield_Current', 'Price_Base', 'Price_Current']
for col in numeric_cols:
  df_data[col] = pd.to_numeric(df_data[col], errors='coerce')
df_data.dropna(subset=numeric_cols, inplace=True)

# 1. 计算总产值指数、总产值增加的绝对额
# 总产值计算
V0 = (df_data['Sown_Area_Base'] * df_data['Yield_Base'] * df_data['Price_Base']).sum()
V1 = (df_data['Sown_Area_Current'] * df_data['Yield_Current'] * df_data['Price_Current']).sum()

total_output_value_index = (V1 / V0) * 100 if V0 != 0 else float('inf')
absolute_increase_total_output_value = V1 - V0

print("\n--- 1. 总产值分析 ---")
print(f"总产值指数: {total_output_value_index:.2f}%")
print(f"总产值增加的绝对额: {absolute_increase_total_output_value}")

# 2. 计算播种面积指数、播种面积增加引起的的产值增加值
A0 = (df_data['Sown_Area_Base'] * df_data['Yield_Base'] * df_data['Price_Base']).sum()
A1 = (df_data['Sown_Area_Current']* df_data['Yield_Base'] * df_data['Price_Base']).sum()

sown_area_index = (A1 / A0) * 100 if A0 != 0 else float('inf')
increase_due_to_sown_area = A1 - A0

print("\n--- 2. 播种面积分析 ---")
print(f"播种面积指数: {sown_area_index:.2f}%")
print(f"播种面积增加引起的产值增加值: {increase_due_to_sown_area}")

# 3. 计算单产指数、单产增加而引起的产值增加额
Y0 = (df_data['Sown_Area_Current'] * df_data['Yield_Base'] * df_data['Price_Base']).sum()
Y1 = (df_data['Sown_Area_Current'] * df_data['Yield_Current'] * df_data['Price_Base']).sum()

yield_per_area_index = (Y1 / Y0) * 100 if Y0 != 0 else float('inf')
increase_due_to_yield_per_area = Y1 - Y0

print("\n--- 3. 单产分析 (Yield Per Unit Area Analysis) ---")
print(f"单产指数: {yield_per_area_index:.2f}%")
print(f"单产增加而引起的产值增加额: {increase_due_to_yield_per_area}")

# 4. 计算粮食价格指数、粮食价格上升增加引起的的产值增加值
P0 = (df_data['Sown_Area_Current'] * df_data['Yield_Current'] * df_data['Price_Base']).sum()
P1 = (df_data['Sown_Area_Current'] * df_data['Yield_Current'] * df_data['Price_Current']).sum()

grain_price_index = (P1 / P0) * 100 if P0 != 0 else float('inf')
increase_due_to_price = P1 - P0

print("\n--- 4. 粮食价格分析 ---")
print(f"粮食价格指数: {grain_price_index:.2f}%")
print(f"粮食价格上升增加引起的产值增加值: {increase_due_to_price}")



```
