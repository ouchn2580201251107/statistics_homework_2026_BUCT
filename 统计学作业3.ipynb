```python
import numpy as np
from scipy.stats import norm

# Given values
sample_mean = 39.5  # 样本平均年龄
sample_size = 36    # 样本数量
pop_std_dev = 7.2   # 总体标准差
confidence_level = 0.99 # 置信水平

# Calculate the Z-score for the given confidence level
# For a two-tailed interval, alpha = 1 - confidence_level
# We need the Z-score corresponding to 1 - alpha/2
alpha = 1 - confidence_level
z_score = norm.ppf(1 - alpha / 2)

# Calculate the standard error of the mean
standard_error = pop_std_dev / np.sqrt(sample_size)

# Calculate the margin of error
margin_of_error = z_score * standard_error

# Calculate the confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

print(f"99% 置信区间: ({lower_bound:.1f}, {upper_bound:.1f})")
```

```python
import numpy as np
from scipy.stats import chi2

# Given values
sample_size = 16    # 样本数量 n
sample_variance = 1.65**2  # 样本方差 S^2
confidence_level = 0.90 # 置信水平

# Calculate degrees of freedom
df = sample_size - 1

# Calculate alpha
alpha = 1 - confidence_level

# Find the Chi-squared critical values
# For lower bound, we use chi2.ppf(1 - alpha/2, df)
# For upper bound, we use chi2.ppf(alpha/2, df)
chi2_lower = chi2.ppf(1 - alpha / 2, df)
chi2_upper = chi2.ppf(alpha / 2, df)

# Calculate the confidence interval for the population variance (sigma^2)
lower_bound_var = (df * sample_variance) / chi2_lower
upper_bound_var = (df * sample_variance) / chi2_upper

print(f"罐头重量方差的90%置信区间: ({lower_bound_var:.4f}, {upper_bound_var:.4f})")
```

```python
import numpy as np
from scipy.stats import norm

# Given values
sample_size = 900    # 样本数量
sample_mean = 1000   # 人均生活费支出 (样本均值)
sample_std_dev = 210 # 标准差 (样本标准差)
confidence_level = 0.9545 # 置信水平

# Calculate alpha
alpha = 1 - confidence_level

# Calculate the Z-score for the given confidence level
# For a two-tailed interval, we need the Z-score corresponding to 1 - alpha/2
z_score = norm.ppf(1 - alpha / 2)

# Calculate the standard error of the mean
standard_error = sample_std_dev / np.sqrt(sample_size)

# Calculate the maximum sampling error (抽样极限误差), which is the margin of error
sampling_limit_error = z_score * standard_error

estimation_precision = 1 - (sampling_limit_error / sample_mean)

# Print the results, rounded to one decimal place
print(f"95.45% 置信水平下的抽样极限误差: {sampling_limit_error:.1f} 元")
print(f"95.45% 置信水平下的抽样估计精度: {estimation_precision * 100 :.1f} %")
```
